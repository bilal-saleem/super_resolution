```python
import sr # Module implemented by me
import matplotlib.pyplot as plt
import cv2
```

```python
eia = sr.loadImages('.\EIA')

# Showing some of the images.
plt.gray()
#plt.figure(figsize=(10, 10))
plt.subplot(121); plt.imshow(eia[0]); plt.title('EIA Low Res')
plt.subplot(122); plt.imshow(eia[-1]); plt.title('EIA Low Res')
plt.show()
```

```python
# First generate parameters needed
# to run the SR algorithm

# F is the affine motion models, based 
Feia = sr.getTranslationModel(eia)
# r is the resolution factor - how much to upscale the images by.
r = 2
size = 3
sigma = 0.4
# H is the point-spread function used 
# within the SR similarity cost function
H = sr.gauss(size, sigma)

# Now, we can run the SR algorithm.
eia_sr = sr.gradient_descent(eia, Feia, H, r)

# Find the cubic spline interpolation
up_dim = (r*eia[0].shape[1], r*eia[0].shape[0])
eia_cubic = cv2.resize(eia[0], (up_dim), interpolation=cv2.INTER_CUBIC)
```

```python
# Showing the results side by side with
# original low-res images.
plt.figure(figsize=(10, 10))
plt.subplot(131); plt.imshow(eia[0])
plt.title('Low-Res EIA')
plt.subplot(132); plt.imshow(eia_cubic)
plt.title('EIA Upscaled, Cubic Interpolation')
plt.subplot(133); plt.imshow(eia_sr)
plt.title('High-Res EIA from Super Resolution')
plt.show()
```

```python
text = sr.loadImages('text')

# Showing some of the images.
plt.gray()
plt.figure(figsize=(5, 5))
plt.subplot(121); plt.imshow(text[0]); plt.title('Low-Res Text')
plt.subplot(122); plt.imshow(text[1]); plt.title('Low-Res Text')
plt.show()
```

```python
Ftext = sr.getTranslationModel(text)
sigma = 0.43
H = sr.gauss(size, sigma)
text_sr = sr.gradient_descent(text, Ftext, H, r)

# Find the cubic spline interpolation
up_dim = (r*text[0].shape[1], r*text[0].shape[0])
text_cubic = cv2.resize(text[0], (up_dim), interpolation=cv2.INTER_CUBIC)

plt.figure(figsize=(5, 5))
plt.subplot(131); plt.imshow(text[0])
plt.title('Low-Res text')
plt.subplot(132); plt.imshow(text_cubic)
plt.title('Text Upscaled, Cubic Interpolation')
plt.subplot(133); plt.imshow(text_sr)
plt.title('High-Res text from \n Super Resolution')
plt.show()
```
